## Class balance

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4))

keys = ['Class 0', 'Class 1']
colors = sns.color_palette('bright')
explode = [0, 0.2]

piechart_cl_train = input_data_train_y.replace({'normal': 0, 'warning': 1}).value_counts()
piechart_cl_test = input_data_test_y.replace({'normal': 0, 'warning': 1}).value_counts()

ax[0].pie(piechart_cl_train, labels=keys, colors=colors, explode=explode, autopct='%.0f%%')
ax[0].set_title('Train dataset')
ax[1].pie(piechart_cl_test, labels=keys, colors=colors, explode=explode, autopct='%.0f%%')
ax[1].set_title('Test dataset');
```

## Which workspace are dangerous vs safer. Which workspaces are in train set and in the test set.

Train set

```python
data_train = pd.concat([input_data_train_X, input_data_train_y], axis=1)
data_train = data_train.rename(columns={0: 'class'})

data_train['class'] = data_train['class'].replace({'normal': 0, 'warning': 1})
```

```python
data_train_grouped_by_id = data_train.groupby('main_working_id')['class'].value_counts().reset_index(name='counts')

#below a list of potentially safe workplaces is created
safe_workplaces = data_train_grouped_by_id['main_working_id'].value_counts()[data_train_grouped_by_id['main_working_id'].value_counts()==1].index.tolist()
safe_workplaces
```

```python
data_train_grouped_by_id
```

```python
data_train[data_train['main_working_id'].isin(safe_workplaces)].iloc[:, 1: 13].describe()
```

```python
data_train[~data_train['main_working_id'].isin(safe_workplaces)].iloc[:, 1: 13].describe()
```

Test set

```python
data_test = pd.concat([input_data_test_X, input_data_test_y], axis=1)
data_test = data_test.rename(columns={0: 'class'})

data_test['class'] = data_test['class'].replace({'normal': 0, 'warning': 1})
```

```python
data_test_grouped_by_id = data_test.groupby('main_working_id')['class'].value_counts().reset_index(name='counts')

#below a list of potentially safe workplaces is created
safe_workplaces = data_test_grouped_by_id['main_working_id'].value_counts()[data_test_grouped_by_id['main_working_id'].value_counts()==1].index.tolist()
```
