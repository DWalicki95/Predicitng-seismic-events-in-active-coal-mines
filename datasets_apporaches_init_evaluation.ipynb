# Datasets testing on 4 chosen algorithms without optimization.

```python
def compare_datasets(X_train, X_test, y_train, y_test):

  model_lr, model_lr_name = LogisticRegression(), 'LogisticRegression'
  model_ada, model_ada_name = AdaBoostClassifier(), 'AdaBoost'
  model_xgb, model_xgb_name = GradientBoostingClassifier(), 'XGBoost'
  model_rf, model_rf_name = RandomForestClassifier(), 'RandomForest'

  models = [model_lr, model_ada, model_xgb, model_rf]
  models_names = [model_lr_name, model_ada_name, model_xgb_name, model_rf_name]
  scores = pd.DataFrame()

  lb = LabelBinarizer()
  y_train_bin = lb.fit_transform(y_train)
  y_train_bin = y_train_bin.ravel()

  y_test_bin = lb.transform(y_test)
  y_test_bin = y_test_bin.ravel()

  for model, model_name in zip(models, models_names):    

    pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', model)
      ])

    pipeline.fit(X_train, y_train_bin)
    y_pred = pipeline.predict(X_test)

    scores.loc['AUC', model_name] = roc_auc_score(y_test_bin, y_pred)
    scores.loc['F1_score', model_name] = f1_score(y_test_bin, y_pred)
    scores.loc['Precision', model_name] = precision_score(y_test_bin, y_pred)
    scores.loc['Recall', model_name] = recall_score(y_test_bin, y_pred)

  return scores
```
